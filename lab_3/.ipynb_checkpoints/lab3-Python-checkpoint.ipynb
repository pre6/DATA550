```python

import pandas as pd

alt.data_transformers.disable_max_rows()
alt.renderers.enable("default")

job_df = pd.read_csv("data/cities-job-satisfaction.csv")

scatter = alt.Chart(job_df).mark_circle(opacity=0.3).encode(
    x=alt.X('last_year:Q', title='last_year'),
    y=alt.Y('this_year:Q', title='this_year')
).properties(
    title="Dramatic increase in dissatisfaction at work",
    height=400,
    width=400
)

scatter

```

<div class="alert alert-info" style="color:black">

### Question 1.3
rubric={viz:2}

<h4>Python</h4>
<p>Improve my Altair plot. Chose the best type of plot to accurately reveal what the data looks like. Design the plot to be an effective plot for communication purposes, this includes proper figure size, axis/figure titles, font sizes, colors, etc.</p>
    
</div>

```python
job_df_updated = job_df[(job_df["last_year"] <= 1) & (job_df["this_year"] <= 1)]
job_df_updated["change"] = job_df["this_year"] - job_df["last_year"]
hist = (
    alt.Chart(job_df_updated)
    .mark_bar(color="gray", stroke="white")
    .encode(
        x=alt.X(
            "change:Q",
            bin=alt.Bin(maxbins=40),
            title="Change in proportion dissatisfied",
            axis=alt.Axis(format="%")
        ),
        y=alt.Y("count()", title="Number of cities"),
    )
)

zero_line = (
    alt.Chart(pd.DataFrame({"x": [0]}))
    .mark_rule(strokeDash=[6, 4], strokeWidth=2)
    .encode(x="x:Q")
)

chart = (
    (hist + zero_line)
    .properties(
        width=650,
        height=450,
        title={
            "text": "Change in job dissatisfaction (this year  last year)",
            "subtitle": "Dashed line = no change; right of 0 means dissatisfaction increased",
            "anchor": "start",
            "fontSize": 20,
            "subtitleFontSize": 13,
        },
    )
    .configure_axis(labelFontSize=12, titleFontSize=14)
)

chart
```

```python
chart2 = alt.Chart(df).mark_bar().encode(
    x=alt.X('median(salary):Q', title='Median salary'), 
    y=alt.Y('when:N', title='When'),                     
    color=alt.Color('when:N', legend=None)               
).properties(
    title='Median of salary',  
    width=500,
    height=200
)

chart2
```

```python
chart3 = alt.Chart(df).mark_bar().encode(
    x=alt.X('count():Q', title='Count'),  
    y=alt.Y('when:N', title='When'),      
    color=alt.Color('when:N', legend=None)
).properties(
    title='Count of Records',  
    width=500,
    height=200
)

chart3
```

### Question 2.3
rubric={viz:2}

<h4>Python</h4>
<p>Improve my Altair plots. Chose the best type of plot to accurately reveal what the data looks like. Design the plot to be an effective plot for communication purposes, this includes proper figure size, axis/figure titles, font sizes, colors, etc. You can either make a single or several plots, depending on what you think is needed.</p>
    
</div>

```python
order = ["after", "before"]
df["when"] = pd.Categorical(df["when"], categories=order, ordered=True)

colors = {"before": "#00BFC4", "after": "#F8766D"}

chart = (
    alt.Chart(df)
    .transform_density(
        density="salary",
        as_=["Salary", "Density"],
        groupby=["when"],
        bandwidth=1.2
    )
    .mark_area(opacity=0.5, line=True)
    .encode(
        x=alt.X("Salary:Q", title="Salary"),
        y=alt.Y("Density:Q", title="Density"),
        color=alt.Color(
            "when:N",
            title="Period",
            sort=order,
            scale=alt.Scale(domain=list(colors.keys()), range=list(colors.values()))
        )
    )
    .properties(
        width=520,
        height=420,
        title="Wage distributions before and after governance"
    )
    .configure_title(fontSize=16, anchor="start")
    .configure_axis(labelFontSize=12, titleFontSize=13)
    .configure_legend(titleFontSize=12, labelFontSize=11)
)

chart
```

```python
df = pd.read_csv('data/guns-r-us.csv')

law_year = 2005

# this was causing me issues for some reason. the data type wasnt numeric and there was a random empty col
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")
df["Firearm murders"] = pd.to_numeric(df["Firearm murders"], errors="coerce")
df = df.dropna(subset=["Year", "Firearm murders"]).copy()
df["zero"] = 0

base = alt.Chart(df).encode(
    x=alt.X("Year:Q", title="Year"),
    y=alt.Y("Firearm murders:Q", title="Firearm murders", scale=alt.Scale(reverse=True))
)

shade = (
    alt.Chart(df)
    .transform_filter(alt.datum.Year)
    .mark_area(color="#9E1B1E", opacity=0.85)
    .encode(
        x="Year:Q",
        y="zero:Q",
        y2="Firearm murders:Q"
    )
)

line = base.mark_line(color="black", strokeWidth=2)
points = base.mark_point(color="black", size=45)

label_df = pd.DataFrame({"Year":[2006], "Firearm murders":[260], "label":["Gun law enacted"]})
label = alt.Chart(label_df).mark_text(
    align="left", dx=10, dy=-5, color="white", fontSize=14
).encode(x="Year:Q", y="Firearm murders:Q", text="label:N")

chart = (shade + line + points + label).properties(width=550, height=420, title="Gun deaths in Florida")
chart
```

```python
line = alt.Chart(df).mark_line(color='darkblue', strokeWidth=2).encode(
    x=alt.X('Year:O', title='Year', axis=alt.Axis(labelFontSize=12, titleFontSize=14)),
    y=alt.Y('Firearm murders:Q', title='Firearm Murders', axis=alt.Axis(labelFontSize=12, titleFontSize=14))
)

points = alt.Chart(df).mark_point(filled=True, size=100, color='red').encode(
    x='Year:O',
    y='Firearm murders:Q',
    tooltip=['Year', 'Firearm murders']  
)

combined_chart = (line + points).properties(
    title=alt.TitleParams('Firearm Murders Over Years', fontSize=16, anchor='middle'),
    width=700,
    height=400
)

combined_chart

```

```python
df2 = df.rename(columns={"the value": "the_value"}).copy()
df2["the_value"] = pd.to_numeric(df2["the_value"], errors="coerce")
alt.data_transformers.disable_max_rows()


chart = (
    alt.Chart(df2)
    .transform_density(
        density="the_value",
        groupby=["where"],
        bandwidth=1600,
        as_=["the_value", "density"],
    )
    .mark_area(opacity=0.45)
    .encode(
        x=alt.X("the_value:Q", title="the value", axis=alt.Axis(format="~s")),
        y=alt.Y("density:Q", title="density", stack=None),
        color=alt.Color("where:N", title=None),
    )
    .properties(title="Ocean waves", width=650, height=420)
)

chart
```

### Question 4.3 (Optional)
rubric={viz:1}

<h4>Python</h4>
<p>Improve my Altair plot. Chose the best type of plot to accurately reveal what the data looks like. Design the plot to be an effective plot for communication purposes, this includes proper figure size, axis/figure titles, font sizes, colors, etc.</p>
    
</div>

```python
summary = (
    df.groupby("where")
      .agg(mean_value=("the value", "mean"),
           n=("the value", "size"))
      .reset_index()
)


import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 5))

bars = ax.bar(
    summary["where"],
    summary["mean_value"],
    color=["#F8766D", "#00BFC4"],
    alpha=0.8
)

# Labels and title
ax.set_ylabel("Mean approval rating")
ax.set_title(
    "Average approval rating by region\n"
    "Bars show means; very small sample sizes",
    fontsize=12
)

# Annotate sample size above bars
for bar, n in zip(bars, summary["n"]):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"n = {n}",
        ha="center",
        va="bottom",
        fontsize=10
    )

ax.set_ylim(0, summary["mean_value"].max() * 1.15)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.show()
```
