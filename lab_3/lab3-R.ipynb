The plot only shows summary statistics (mean, median), which says nothing about how individual salaries changed.
A higher mean does not imply that most workers earned more. It could be driven by a small number of very high salaries (right-skew).

```r
# YOUR ANSWER HERE

df <- read.csv("data/wages.csv")


# Mean salaries
mean_df <- df %>%
  group_by(when) %>%
  summarise(value = mean(salary), .groups = "drop") %>%
  mutate(stat = "Mean salary")

# Median salaries
median_df <- df %>%
  group_by(when) %>%
  summarise(value = median(salary), .groups = "drop") %>%
  mutate(stat = "Median salary")

# Counts
count_df <- df %>%
  count(when, name = "value") %>%
  mutate(stat = "Count")
```

```r
plot_df <- bind_rows(mean_df, median_df, count_df)
```

```r
ggplot(plot_df, aes(x = value, y = when, fill = when)) +
  geom_col(width = 0.6) +
  facet_wrap(~ stat, scales = "free_x", ncol = 1) +
  scale_fill_manual(values = c("before" = "#00BFC4", "after" = "#F8766D")) +
  scale_x_continuous(labels = dollar_format(prefix = "$")) +
  labs(
    x = NULL,
    y = "when"
  ) +
  theme_gray(base_size = 13) +
  theme(
    legend.position = "none",
    strip.text = element_text(face = "bold"),
    panel.grid.minor = element_blank()
  )
```
